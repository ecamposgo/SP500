```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd                                
pd.set_option('display.max_columns', None)

from sklearn.preprocessing import StandardScaler, MinMaxScaler 
```

```python
#fichero con datos historicos de todos los sectores
historical_sectors = pd.read_excel('../data/historical_sectors.xlsx')
```

```python
historical_sectors=historical_sectors.drop(['Adj Close','Volume'], axis=1)
```

```python
#escalar la columna de precio de cierre
data=historical_sectors.filter(['Close'])
dataset = data.values

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(dataset)

historical_sectors['Scaled_Close'] = scaled_data
```

```python
historical_sectors.columns= historical_sectors.columns.str.lower()
historical_sectors = historical_sectors.dropna()
historical_sectors=historical_sectors[['ticker','date','sector','open','high','low','close','scaled_close']]
```

```python
#calculo el porcentaje de crecimiento
historical_sectors['return']=round((historical_sectors['close']/historical_sectors['open']-1)*100,2)
historical_sectors.head()
```

```python
#fichero con datos historicos sp500
historical_sp500 = pd.read_excel('../data/historical_sp500.xlsx')
```

```python
historical_sp500=historical_sp500.drop(['Adj Close','Volume'], axis=1)
```

```python
#escalar la columna de precio de cierre
data=historical_sp500.filter(['Close'])
dataset = data.values

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(dataset)

historical_sp500['Scaled_Close'] = scaled_data
```

```python
historical_sp500.columns= historical_sp500.columns.str.lower()
historical_sp500 = historical_sp500.dropna()
historical_sp500['sector']='SP500'
historical_sp500=historical_sp500[['ticker','date','sector','open','high','low','close','scaled_close']]
historical_sp500
```

```python
#calculo el porcentaje de crecimiento
historical_sp500['return']=round((historical_sp500['close']/historical_sp500['open']-1)*100,2)
historical_sp500.head()
```

```python
#concatenar sectores + sp500
historical_sp500_sectors=pd.concat([historical_sectors, historical_sp500])

historical_sp500_sectors.to_excel('../data/historical_sp500_sectors_clean.xlsx',index=False)
```
