```python
import pandas as pd


import warnings
warnings.filterwarnings('ignore')

import pandas as pd                                
pd.set_option('display.max_columns', None)

import yfinance as yf
```

```python
#funcion para descargar datos 

def get_data(Symbol):
    
    data=yf.download(tickers=Symbol, start='2000-01-01', end='2022-10-01', interval='1mo')
    data['ticker']=Symbol
    data['date']=data.index
  
    return data
```

```python
#descargar datos historicos S&P 500
historical_sp500=get_data('^GSPC')
```

```python
#guardar datos historicos del sp500
historical_sp500.to_excel('../data/historical_sp500.xlsx',index=False)
```

```python
#descargar datos historicos S&P 500 por sectores

list_of_symbols = ['^GSPE','^SP500-15','^SP500-20','^SP500-25','^SP500-30','^SP500-35','^SP500-40','^SP500-45',
                  '^SP500-50','^SP500-55','^SP500-60']

historical_sectors = pd.DataFrame(columns=['Open','High','Low','Close','Adj Close','Volume'])

for symbol in list_of_symbols:
    df=get_data(symbol)
    historical_sectors = pd.concat([historical_sectors, df])
```

```python
historical_sectors['Sector'] = ''
historical_sectors.loc[historical_sectors['ticker'] == '^GSPE', 'Sector'] = 'Energy'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-15', 'Sector'] = 'Materials'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-20', 'Sector'] = 'Industrials'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-25', 'Sector'] = 'Consumer Discretionary'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-30', 'Sector'] = 'Consumer Staples'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-35', 'Sector'] = 'Health Care'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-40', 'Sector'] = 'Financials'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-45', 'Sector'] = 'Information Technology'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-50', 'Sector'] = 'Communication Services'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-55', 'Sector'] = 'Utilities'
historical_sectors.loc[historical_sectors['ticker'] == '^SP500-60', 'Sector'] = 'Real State'
```

```python
#guardar datos historicos del sp500
historical_sectors.to_excel('../data/historical_sectors.xlsx',index=False)
```

```python
#descarga datos historicos empresas 

companies = pd.read_excel('../data/S&P500_companies.xlsx')

list_of_symbols = companies['Symbol'].tolist()

historical_companies = pd.DataFrame(columns=['Open','High','Low','Close','Adj Close','Volume'])


for symbol in list_of_symbols:
    historical_companies = pd.concat([historical_companies, get_data(symbol)])

```

```python
#guardo csv de copia
historical_companies.to_excel('../data/historical_companies.xlsx',index=False)
```
